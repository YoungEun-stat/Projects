```python
############# bike train pre #############
# split / split2 하나의 함수로 만들기!! (코드 정리)

def split(data) :
    
    data['년'] = data.대여일시.str.split('.').str[0].astype(int)
    data['월'] = data.대여일시.str.split('.').str[1].astype(int)
    data['일'] = data.대여일시.str.split('.').str[2].astype(int)

    data = data.drop(['대여일시'],axis=1)

    return data

def split2(data) :
    
    data['년'] = data.대여일시.str.split('-').str[0].astype(int)
    data['월'] = data.대여일시.str.split('-').str[1].astype(int)
    data['일'] = data.대여일시.str.split('-').str[2].astype(int)

    data = data.drop(['대여일시'],axis=1)

    return data
```

```python
bike_1906_1911 = split(bike_1906_1911) # 2019_6 ~ 2019_11 # 1907907 : NA
bike_1912_2005 = split(bike_1912_2005) # 2019_12 ~2020_05 # 
bike_2006 = split(bike_2006) # 2020_06
bike_2007_2101 = split2(bike_2007_2101) # 2020_07 ~ 2021_01
bike_2102_2106 = split2(bike_2102_2106) # 2021_02 ~ 2021_06
```

```python
# bike_1912_2005 일부분만 가지고 오기 (2019_12 ~2020_05)
bike_1912 =  bike_1912_2005[bike_1912_2005.월 == 12]
bike_2001_2005 =  bike_1912_2005[bike_1912_2005.년 == 2020]
```
